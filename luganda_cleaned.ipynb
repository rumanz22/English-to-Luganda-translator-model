Overview of the Project: English to Luganda Translator

The English to Luganda Translator is a machine translation application designed to bridge the gap between English and Luganda, a widely spoken language in Uganda. It leverages the power of the MarianMT model by Hugging Face, a state-of-the-art neural machine translation framework, to provide accurate and context-aware translations. This project aims to showcase how AI can facilitate communication and preserve cultural heritage through language technology.

Key Objectives:

Build an efficient translation system for English to Luganda.

Enable interactive demonstrations for both technical and non-technical users.

Provide an intuitive, user-friendly interface powered by Gradio.

Process of Creating the Project
1. Research and Planning
Identified Problem: A lack of accessible tools for translating English to Luganda.
Solution: Use the pre-trained MarianMT model from Hugging Face, specifically fine-tuned for English-to-Luganda translation.

2. Dataset Preparation
Source Data: A dataset (Luganda.csv) containing English sentences and their corresponding Luganda translations was used.

Preprocessing:
Checked for missing values.
Selected relevant columns (English and Luganda).
Renamed columns for consistency (source for English, target for Luganda).

3. Model Setup
Model Selection: The MarianMT model (Helsinki-NLP/opus-mt-en-lg) was chosen for its effectiveness in low-resource language translation.

Tokenizer and Model Loading:
Tokenizer prepares input sentences for the model.
MarianMT generates translations based on trained parameters.

4. Application Development
Gradio Interface: Designed a web-based interface that:
Accepts English input for translation.
Displays the Luganda translation output.
Provides sample translations using examples from the dataset.

5. Evaluation
Testing: Sampled sentences from the dataset to compare model predictions with reference translations (ground truth).
Quality Assessment: Used BLEU scores to evaluate translation accuracy and identify areas for improvement.

6. Deployment
The project is deployed locally and can be hosted online using platforms like Hugging Face Spaces, Streamlit Cloud, or AWS.

```python
import gradio as gr
from transformers import MarianTokenizer, MarianMTModel

# Load model and tokenizer
model_name = "Helsinki-NLP/opus-mt-en-lg"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_to_luganda(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)
    outputs = model.generate(inputs["input_ids"])
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Create Gradio interface
interface = gr.Interface(
    fn=translate_to_luganda,
    inputs="text",
    outputs="text",
    title="English to Luganda Translator"
)

# Launch interface
interface.launch()
```

```python
import gradio as gr
import pandas as pd
from transformers import MarianTokenizer, MarianMTModel

# Load model and tokenizer
model_name = "Helsinki-NLP/opus-mt-en-lg"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_to_luganda(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)
    outputs = model.generate(inputs["input_ids"])
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Load the dataset
dataset_path = "Luganda.csv"  # Update the path if needed
try:
    df = pd.read_csv(dataset_path, encoding="latin1")
    df = df[['English', 'Luganda']].dropna()  # Ensure relevant columns and drop missing values
    df = df.rename(columns={"English": "source", "Luganda": "target"})
except Exception as e:
    df = None
    print(f"Error loading dataset: {e}")

# Function to provide example translations from the dataset
def example_translation(index):
    if df is not None and 0 <= index < len(df):
        row = df.iloc[index]
        source = row['source']
        target = row['target']
        prediction = translate_to_luganda(source)
        return f"Input: {source}\n\nPredicted: {prediction}\n\nGround Truth: {target}"
    return "Example index out of range or dataset not loaded."

# Create Gradio interface
with gr.Blocks() as interface:
    gr.Markdown("# English to Luganda Translator")

    with gr.Tab("Translate Sentence"):
        input_text = gr.Textbox(label="Enter an English sentence")
        output_text = gr.Textbox(label="Translated Luganda sentence")
        translate_button = gr.Button("Translate")

        def handle_translation(sentence):
            return translate_to_luganda(sentence)

        translate_button.click(handle_translation, inputs=input_text, outputs=output_text)

    with gr.Tab("Example Translations"):
        example_index = gr.Slider(minimum=0, maximum=len(df) - 1 if df is not None else 0, step=1, label="Select Example Index")
        example_output = gr.Textbox(label="Example Translation Output")
        example_button = gr.Button("Show Example")

        def show_example(index):
            return example_translation(index)

        example_button.click(show_example, inputs=example_index, outputs=example_output)

# Launch interface
interface.launch()
```
